ðŸ§© CELL 1: Import Libraries & Load Clean Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Better visuals
sns.set(style="whitegrid")

# Load cleaned data (from Phase 2)
df = pd.read_csv("../data/processed/cleaned_burnout_data.csv")

df.head()
```

ðŸ§© CELL 2: Burnout Risk Distribution (MOST IMPORTANT)

```python
df['Burnout_Risk'].value_counts()
```

```python
df['Burnout_Risk'].value_counts().plot(
    kind='bar',
    title='Burnout Risk Distribution',
    xlabel='Burnout Risk Level',
    ylabel='Number of Employees'
)
plt.show()
```

ðŸ§© CELL 3: Burnout vs Overtime

```python
sns.countplot(x='Burnout_Risk', hue='OverTime', data=df)
plt.title("Burnout Risk vs Overtime")
plt.xlabel("Burnout Risk")
plt.ylabel("Employee Count")
plt.show()
```

ðŸ§© CELL 4: Work-Life Balance vs Burnout

```python
sns.boxplot(x='Burnout_Risk', y='WorkLifeBalance', data=df)
plt.title("Work-Life Balance vs Burnout Risk")
plt.show()
```

ðŸ§© CELL 5: Job Satisfaction vs Burnout

```python
sns.boxplot(x='Burnout_Risk', y='JobSatisfaction', data=df)
plt.title("Job Satisfaction vs Burnout Risk")
plt.show()
```

ðŸ§© CELL 6: Environment Satisfaction vs Burnout

```python
sns.boxplot(x='Burnout_Risk', y='EnvironmentSatisfaction', data=df)
plt.title("Environment Satisfaction vs Burnout Risk")
plt.show()
```

 Burnout Risk by Age Group

```python
# Create age groups
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[18, 30, 40, 50, 60],
    labels=['18-30', '31-40', '41-50', '51-60']
)

sns.countplot(x='AgeGroup', hue='Burnout_Risk', data=df)
plt.title("Burnout Risk by Age Group")
plt.show()
```

Burnout Risk vs Monthly Income

```python
sns.boxplot(x='Burnout_Risk', y='MonthlyIncome', data=df)
plt.title("Monthly Income vs Burnout Risk")
plt.show()
```

Burnout Risk vs Years at Company

```python
sns.boxplot(x='Burnout_Risk', y='YearsAtCompany', data=df)
plt.title("Years at Company vs Burnout Risk")
plt.show()
```

Heatmap of Burnout-Related Features

```python
numeric_cols = [
    'Age',
    'MonthlyIncome',
    'YearsAtCompany',
    'WorkLifeBalance',
    'JobSatisfaction',
    'EnvironmentSatisfaction',
    'PerformanceRating',
    'OverTime'
]

plt.figure(figsize=(10,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Burnout-Related Features")
plt.show()
```
